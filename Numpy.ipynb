#### Built In Methods

#### arange 

Returns evenly spaced values in a given interval

```python
np.arange(0,10)
```

```python
np.arange(0,11,2)
```

#### zeros and ones

Generates arrays of zeros and ones

```python
np.zeros(3)
```

```python
np.zeros((3,3))
```

```python
np.ones(4)
```

```python
np.ones((4,4))
```

#### linspace

Returns evenly spaced numbers over a specified interval

```python
np.linspace(0,10,50)
```

#### Random

#### rand
Creates array of given shape and populates it with random samples from a uniform distribution over 0 - 1.

```python
np.random.rand(10)
```

#### randn

Returns a sample from normal distribution

```python
np.random.randn(2)
```

```python
np.random.randn(5,5)
```

#### randint

Return random integers from low to high

```python
np.random.randint(1,100)
```

```python
np.random.randint(1,100,10)
```
