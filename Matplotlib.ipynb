#### Grandfather of visualization libraries

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,5,11)

y = x**2
```

```python
plt.plot(x,y,'r')
plt.xlabel('X Axis title here')
plt.ylabel('Y Axis title here')
plt.title('Graph title here')
```

```python
plt.subplot(1,3,1) # nrows, ncols, plot_number
plt.plot(x,y);

plt.subplot(1,3,2) # nrows, ncols, plot_number
plt.plot(y,x,'r--');

plt.subplot(1,3,3) # nrows, ncols, plot_number
plt.plot(x,y,'g+');
```

```python
fig = plt.figure(figsize=(10,5))

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width and height

axes.plot(x,y,'b',label='x**2')
axes.plot(y,x+y,'r',label='x+y')
axes.legend(loc=4)

# axes.set_xlabel
# axes.set_ylabel
```

```python
fig,ax = plt.subplots()#1,2,figsize=(12,5))

ax.plot(x,x**2,color='blue',alpha=0.5,linewidth=6.5,linestyle='-.',marker='+',
       markersize=9)
ax.plot(x,x**3,color='#493E63',linewidth=9.5,linestyle='-',marker='o',
       markersize=10,markerfacecolor='yellow',markeredgecolor='blue',markeredgewidth=3)
```
